# Understanding Ploting using pandas plot 
- sharex will label years(2000-2020) on x-axis for each subplot(Open, High, Low, Close, Adj Close and Volume)
  instead of sharing it among all by mentioning it once at end subplots(Close, Adj Close and Volume) 
- rot will rotate x axis values 2000 - 2020 by 90 degree vertically downwards

```python
column_list = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
df.plot(x = 'Date',y = column_list, subplots = True, layout = (3, 3),
        figsize = (15, 15), sharex = False, title = "Stock Value Trends over Time", rot = 90)
```

# Finding Correlation and visulaizing using seaborn
- vmax, vmin :
    Anchoring the colormap with setting vmax and vmin, color bar on side too changes as per the range.
    Also if there are multiple data to be plotted , data between the range will only be displayed.

- cmap : Sequential colormap

- annot : values of corr dispalyed in ceter

- fmt : values of corr displayed till that decimal place( example here upto 3)

- xticklables/ yticklabels if false removes that labels

```python
sns.heatmap(df.corr(), vmin=0, vmax=1, cmap='YlGnBu', annot=True, fmt='.3g',yticklabels=False,
            cbar_kws={'label': 'My Colorbar', 'orientation': 'horizontal'})
```

# Undering Outlier using box plot

- To create empty figure boxes:
    - In this case 6 boxes as len(column_list) = 6 and axes (rectangle with o.o to 1.0 on both x and y axis):
         - fig, ax = plt.subplots (len(column_list), figsize = (10, 20)) 
- for i, col_list in enumerate(column_list): Printing i and col_list : \
  0 Open \
  1 High \
  2 Low \
  3 Close \
  4 Adj Close \
  5 Volume 
- sns.boxplot(df[col_list],ax=ax[i], palette = "winter", orient = 'h') \
    - Here ax = ax[i] will plot individual index column data subplot, if not mentioned only volume subplot will be seen
    
- ax[i].set_title("Whisker Plot for Outlier Detection on" + " " + col_list, fontsize = 10)
- ax[i].set_ylabel(col_list, fontsize = 8)
    - setting x and y axis label
    
- fig.tight_layout(pad = 1.1)
    - padding between two subplot boxes/ rectangles

```python
fig, ax = plt.subplots (len(column_list), figsize = (10, 20)) 
for i, col_list in enumerate(column_list):
    sns.boxplot(df[col_list],ax=ax[i], palette = "winter", orient = 'h')
    ax[i].set_title("Whisker Plot for Outlier Detection on" + " " + col_list, fontsize = 10)
    ax[i].set_ylabel(col_list, fontsize = 8)
    fig.tight_layout(pad = 1.1)
```
